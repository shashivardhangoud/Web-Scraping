```python
import requests
wiki_link="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area"
links=requests.get(wiki_link).text
```

```python
print(links)
```

```python
from bs4 import BeautifulSoup
sp=BeautifulSoup(links,'lxml')
```

```python
print(sp)
```

```python
print(sp.prettify())
```

```python
sp.title
```

```python
sp.find_all("a")
```

```python

all_link=sp.find_all("a")
for link in all_link:
    print(link.get("href"))
```

```python
all_tables=sp.find_all("table")
print(all_tables)
```

```python
right_table=sp.find('table',class_="wikitable sortable")
print(right_table)
```

```python
table_links=right_table.find_all("a")
table_links
```

```python
country=[]
for links in table_links:
    country.append(links.get('title'))
```

```python
print(country)
```
